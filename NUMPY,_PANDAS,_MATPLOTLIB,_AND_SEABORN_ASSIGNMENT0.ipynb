## Pandas Operations

This section will demonstrate fundamental Pandas operations, including selecting data, filtering rows, handling missing values, and grouping data.

### 1. Select specific columns

```python
# Select a single column
age_column = df['Age']
print("Age column (first 5 rows):\n", age_column.head())

# Select multiple columns
subset_df = df[['Name', 'Age', 'Pclass', 'Fare']]
print("\nSubset DataFrame (first 5 rows):\n")
display(subset_df.head())
```

### 2. Filter rows based on conditions

```python
# Filter for passengers who survived
survived_passengers = df[df['Survived'] == 1]
print("Survivors (first 5 rows):\n")
display(survived_passengers.head())

# Filter for female passengers older than 30
female_older_than_30 = df[(df['Sex'] == 'female') & (df['Age'] > 30)]
print("\nFemale passengers older than 30 (first 5 rows):\n")
display(female_older_than_30.head())
```

### 3. Handle missing values

```python
# Check for missing values
print("Missing values before handling:\n", df.isnull().sum())

# Fill missing 'Age' values with the mean
df_filled_age = df.copy()
df_filled_age['Age'].fillna(df_filled_age['Age'].mean(), inplace=True)
print("\nMissing 'Age' values after filling with mean:\n", df_filled_age['Age'].isnull().sum())

# Drop rows with any missing values (e.g., 'Cabin' has many NaNs)
df_dropped_na = df.dropna()
print("\nShape after dropping rows with any missing values:", df_dropped_na.shape)

# Fill missing 'Embarked' with the mode
df_filled_embarked = df.copy()
mode_embarked = df_filled_embarked['Embarked'].mode()[0]
df_filled_embarked['Embarked'].fillna(mode_embarked, inplace=True)
print("\nMissing 'Embarked' values after filling with mode:\n", df_filled_embarked['Embarked'].isnull().sum())
```

### 4. Group data using `groupby()`

```python
# Group by 'Sex' and calculate the mean of 'Age' and 'Fare'
grouped_sex = df.groupby('Sex')[['Age', 'Fare', 'Survived']].mean()
print("Mean Age, Fare, and Survived by Sex:\n")
display(grouped_sex)

# Group by 'Pclass' and count the number of passengers
grouped_pclass = df.groupby('Pclass').size().reset_index(name='Count')
print("\nPassenger count by Pclass:\n")
display(grouped_pclass)
```

## NumPy Operations

This section will demonstrate fundamental NumPy operations, including array creation, element-wise operations, mathematical functions, array reshaping, and array concatenation.

### 1. Create arrays using `np.array()`, `np.zeros()`, `np.ones()`, `np.random.rand()`

```python
import numpy as np

# Using np.array()
arr_from_list = np.array([1, 2, 3, 4, 5])
print("Array from list:", arr_from_list)

# Using np.zeros()
zeros_array = np.zeros((3, 4))
print("\nZeros array (3x4):\n", zeros_array)

# Using np.ones()
ones_array = np.ones((2, 3))
print("\nOnes array (2x3):\n", ones_array)

# Using np.random.rand()
random_array = np.random.rand(2, 2)
print("\nRandom array (2x2):\n", random_array)
```

### 2. Perform element-wise operations and apply mathematical functions

```python
# Element-wise operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Array a:", a)
print("Array b:", b)
print("\nAddition (a + b):", a + b)
print("Subtraction (a - b):", a - b)
print("Multiplication (a * b):", a * b)
print("Division (a / b):", a / b)

# Apply mathematical functions
x = np.array([0, np.pi/2, np.pi])
print("\nArray x:", x)
print("Sine of x (np.sin(x)):", np.sin(x))
print("Cosine of x (np.cos(x)):", np.cos(x))
print("Square root of b (np.sqrt(b)):", np.sqrt(b))
```

### 3. Reshape arrays using `.reshape()`

```python
original_array = np.arange(1, 10) # Creates an array [1, 2, ..., 9]
print("Original array:", original_array)
print("Original shape:", original_array.shape)

reshaped_array = original_array.reshape(3, 3)
print("\nReshaped array (3x3):\n", reshaped_array)
print("Reshaped shape:", reshaped_array.shape)

# Reshape to a 1D array
flat_array = reshaped_array.reshape(-1)
print("\nFlattened array:", flat_array)
```

### 4. Concatenate arrays using `np.hstack()` and `np.vstack()`

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

print("Array 1:\n", arr1)
print("\nArray 2:\n", arr2)

# Horizontal stack (hstack)
h_stacked_array = np.hstack((arr1, arr2))
print("\nHorizontally stacked array (np.hstack):\n", h_stacked_array)

# Vertical stack (vstack)
v_stacked_array = np.vstack((arr1, arr2))
print("\nVertically stacked array (np.vstack):\n", v_stacked_array)
```

## Optimize Data Types for Memory Reduction

### Subtask:
Inspect the current memory usage and data types of the `tips_df` DataFrame. Identify columns where data types can be optimized (e.g., converting float64 to float32, int64 to smaller integer types, or object to category) using `astype()` to reduce memory footprint. Display the memory usage before and after optimization.


**Reasoning**:
The subtask requires inspecting memory usage before optimization, creating an optimized DataFrame with reduced data types for specified columns, and then displaying memory usage after optimization. I will use `memory_usage(deep=True)` for accurate memory calculation and `astype()` for type conversions.



```python
import pandas as pd

# Reload tips_df from original source to ensure we start with the original data types
# (It is confirmed to be present in the kernel state: '/kaggle/input/restaurant-tips-dataset/tips.csv')
tips_df = pd.read_csv(tips_csv_path)

print("Memory usage before optimization:")
print(tips_df.memory_usage(deep=True))
print(f"Total memory usage before optimization: {tips_df.memory_usage(deep=True).sum() / (1024*1024):.2f} MB\n")

# Create a deep copy for optimization
tips_df_optimized = tips_df.copy(deep=True)

# Convert 'total_bill' and 'tip' from float64 to float32
tips_df_optimized['total_bill'] = tips_df_optimized['total_bill'].astype('float32')
tips_df_optimized['tip'] = tips_df_optimized['tip'].astype('float32')

# Convert 'size' from int64 to int8
tips_df_optimized['size'] = tips_df_optimized['size'].astype('int8')

# Convert categorical columns from object to category
for col in ['sex', 'smoker', 'day', 'time']:
    tips_df_optimized[col] = tips_df_optimized[col].astype('category')

print("Memory usage after optimization:")
print(tips_df_optimized.memory_usage(deep=True))
print(f"Total memory usage after optimization: {tips_df_optimized.memory_usage(deep=True).sum() / (1024*1024):.2f} MB")
```
